# **2. Line Plots**
Best for: Time series data, trends over continuous intervals

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample dataset: Monthly temperature data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature_2023 = [5, 7, 12, 16, 20, 25, 28, 27, 22, 17, 11, 6]
temperature_2024 = [6, 8, 13, 15, 21, 26, 29, 28, 23, 16, 12, 7]

plt.figure(figsize=(12, 6))
plt.plot(months, temperature_2023, marker='o', linewidth=2, label='2023')
plt.plot(months, temperature_2024, marker='s', linewidth=2, label='2024')
plt.title('Monthly Temperature Comparison (2023 vs 2024)')
plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

This line plot compares the monthly temperatures in 2023 and 2024. You can see the temperature trends over the months for both years. The plot shows that temperatures generally increase from January to July/August and then decrease towards December in both years. You can also compare the temperatures between the two years for each month. For example, you can see that July 2024 was slightly warmer than July 2023.

# **3. Scatter Plots**
Best for: Relationship between two numerical variables

```
# Sample dataset: Student performance
np.random.seed(42)
study_hours = np.random.normal(15, 5, 100)
exam_scores = 50 + 2*study_hours + np.random.normal(0, 10, 100)

plt.figure(figsize=(10, 6))
plt.scatter(study_hours, exam_scores, alpha=0.7, c='blue', edgecolors='black')
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours per Week')
plt.ylabel('Exam Score')
plt.grid(True, alpha=0.3)

# Add trend line
z = np.polyfit(study_hours, exam_scores, 1)
p = np.poly1d(z)
plt.plot(study_hours, p(study_hours), "r--", alpha=0.8, label=f'Trend: y={z[0]:.1f}x+{z[1]:.1f}')
plt.legend()
plt.show()
```

np.random.seed(42): This line sets the "seed" for the random number generator. This is done so that every time you run this code, you get the same sequence of "random" numbers. This is useful for making your results reproducible.

study_hours = np.random.normal(15, 5, 100): This line generates 100 random numbers that follow a normal distribution. The numbers are centered around 15 (the mean) and have a standard deviation of 5.
These numbers are stored in the variable study_hours and are intended to represent the number of hours students study per week.


exam_scores = 50 + 2*study_hours + np.random.normal(0, 10, 100): This line calculates the exam_scores. It starts with a base score of 50.

Then, for each student, it adds twice their study_hours to this base score.

Finally, it adds some random noise (generated from a normal distribution with a mean of 0 and a standard deviation of 10) to each score.

This simulates the idea that exam scores are influenced by study hours but also have some variability due to other factors.


This scatter plot shows the relationship between the number of hours students study per week and their exam scores.

Each point on the graph represents one student.
The position of the point on the horizontal axis (x-axis) indicates the number of hours that student studied.
The position of the point on the vertical axis (y-axis) indicates the score that student received on the exam.
The red dashed line is a trend line, showing the general direction of the relationship between study hours and exam scores. The equation y = 2.1x + 50.0 indicates that, on average, for every additional hour studied, the exam score increases by approximately 2.1 points, starting from a base score around 50.
From the plot, you can observe a positive correlation between study hours and exam scores – as study hours increase, exam scores tend to increase as well. The points are not perfectly on the line, which is expected as other factors besides study hours can influence exam scores.



plt.plot(study_hours, p(study_hours), "r--", alpha=0.8, label=f'Trend: y={z[0]:.1f}x+{z[1]:.1f}'): This line adds a trend line to the scatter plot.

study_hours: These are the x-values for the line.

p(study_hours): This calculates the corresponding y-values for the trend line using the polynomial p that was fitted to the data.

"r--": This specifies the style of the line. "r" means red and "--" means a dashed line.

alpha=0.8: This sets the transparency of the line, making it slightly see-through.

label=f'Trend: y={z[0]:.1f}x+{z[1]:.1f}': This sets the label for the

trend line that will appear in the legend. The f'...' is a formatted

string literal that includes the calculated slope (z[0]) and intercept (z[1]) of the trend line, formatted to one decimal place.


plt.legend(): This line displays the legend on the plot, which includes the label for the trend line.

plt.show(): This line renders and displays the plot.

# **4. Visualizing Errors**
Best for: Showing uncertainty in measurements

```
# Sample dataset: Experimental measurements with errors
x = np.linspace(0, 10, 20)
y = 2*x + 1 + np.random.normal(0, 2, 20)
y_error = np.random.uniform(0.5, 2, 20)  # Varying error magnitudes

plt.figure(figsize=(12, 5))
```

```
# Subplot 1: Error bars
plt.subplot(1, 2, 1)
plt.errorbar(x, y, yerr=y_error, fmt='o', color='blue',
             ecolor='red', elinewidth=2, capsize=5, alpha=0.7)
plt.title('Error Bars: Measurements with Uncertainty')
plt.xlabel('X values')
plt.ylabel('Y values with error')
plt.grid(True, alpha=0.3)
```

# **Left Subplot (Error Bars): This plot uses error bars to show the uncertainty in each data point.**
The blue circles represent the measured values (y).
The vertical red lines extending above and below each blue circle are the error bars. Their length indicates the magnitude of the uncertainty (y_error) associated with that particular measurement. A longer error bar means more uncertainty.
The plot titled 'Error Bars: Measurements with Uncertainty' visually communicates that each measurement has a range of possible true values, not just a single point.


```
# Subplot 2: Confidence interval
plt.subplot(1, 2, 2)
plt.plot(x, y, 'o-', color='blue', alpha=0.7, label='Mean values')
plt.fill_between(x, y-y_error, y+y_error, alpha=0.3, color='red', label='Confidence interval')
plt.title('Confidence Intervals')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

# **Right Subplot (Confidence Intervals): **

This plot uses a shaded area to represent the confidence interval around the data points.

The blue line with circles shows the mean or central tendency of the measurements (y).

The shaded red area around the blue line represents the confidence interval.

This area indicates a range within which the true value is likely to fall, based on the data and the calculated error (y_error).

The plot titled 'Confidence Intervals' provides a visual band of uncertainty, suggesting a likely range for the underlying relationship or trend in the data, rather than focusing on the uncertainty of individual points.


Both plots effectively show the variability and uncertainty in the measurements, but they do so in slightly different ways: error bars emphasize the uncertainty of individual points, while confidence intervals highlight the likely range for the overall trend.

# **5. Density and Contour Plots**

Best for: Visualizing 3D data in 2D, probability distributions

```
from scipy.stats import gaussian_kde

# Sample dataset: Bivariate normal distribution
np.random.seed(42)
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000) + 0.5*x

plt.figure(figsize=(15, 5))
```

```
# Subplot 1: 2D Density plot
plt.subplot(1, 3, 1)
# Calculate point density
xy = np.vstack([x, y])
z = gaussian_kde(xy)(xy)
plt.scatter(x, y, c=z, s=50, cmap='viridis')
plt.colorbar(label='Density')
plt.title('2D Density Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
```

Left Subplot (2D Density Scatter Plot): This plot shows the individual data points, but the color of each point indicates the density of points in that area.
The color bar on the right shows that warmer colors (like yellow) represent higher density, while cooler colors (like purple) represent lower density.
You can see where the data points are clustered more tightly by looking at the areas with warmer colors.

```
# Subplot 2: Contour plot
plt.subplot(1, 3, 2)
# Create grid for contour plot
x_grid = np.linspace(x.min(), x.max(), 100)
y_grid = np.linspace(y.min(), y.max(), 100)
X, Y = np.meshgrid(x_grid, y_grid)
positions = np.vstack([X.ravel(), Y.ravel()])
Z = gaussian_kde(xy)(positions)
Z = Z.reshape(X.shape)

contour = plt.contour(X, Y, Z, levels=8, colors='black')
plt.clabel(contour, inline=True, fontsize=8)
plt.imshow(Z, extent=[x.min(), x.max(), y.min(), y.max()],
           origin='lower', cmap='viridis', alpha=0.5)
plt.title('Contour Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
```

Middle Subplot (Contour Plot): This plot uses contour lines to show areas of equal density, similar to contour lines on a topographic map showing equal elevation.
Each line represents a specific density level.
Where the lines are closer together, the density is changing rapidly. Where they are further apart, the density is changing more slowly.
The numbers on the lines indicate the density value for that contour level.

```
# Subplot 3: Filled contour plot
plt.subplot(1, 3, 3)
plt.contourf(X, Y, Z, levels=15, cmap='viridis')
plt.colorbar(label='Density')
plt.title('Filled Contour Plot')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.tight_layout()
plt.show()
```

Right Subplot (Filled Contour Plot): This plot is similar to the contour plot but fills the areas between the contour lines with color.
This provides a more intuitive visual representation of the density distribution.
Again, warmer colors indicate higher density, and cooler colors indicate lower density, as shown by the color bar.
All three plots help to visualize the distribution of the data in two dimensions, highlighting where the data is most concentrated.

# **6. Histograms**
Best for: Understanding distribution of numerical data

```
# Sample dataset: Exam scores from three different classes
np.random.seed(42)
class_a = np.random.normal(75, 10, 200)
class_b = np.random.normal(65, 15, 200)
class_c = np.random.normal(85, 8, 200)

plt.figure(figsize=(15, 5))
```

```
# Subplot 1: Basic histogram
plt.subplot(1, 3, 1)
plt.hist(class_a, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title('Single Distribution: Class A Scores')
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
```

```
# Subplot 2: Multiple histograms
plt.subplot(1, 3, 2)
plt.hist(class_a, bins=20, alpha=0.5, label='Class A', color='blue')
plt.hist(class_b, bins=20, alpha=0.5, label='Class B', color='red')
plt.hist(class_c, bins=20, alpha=0.5, label='Class C', color='green')
plt.title('Multiple Distributions Comparison')
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)
```

```
# Subplot 3: Density histogram
plt.subplot(1, 3, 3)
plt.hist(class_a, bins=20, density=True, alpha=0.7, color='blue',
         edgecolor='black', label='Normalized')
# Add a normal curve for comparison
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, class_a.mean(), class_a.std())
plt.plot(x, p, 'k', linewidth=2, label='Normal distribution')
plt.title('Density Histogram with Normal Curve')
plt.xlabel('Exam Scores')
plt.ylabel('Density')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

# **7. Comparative Analysis with Same Dataset**
Let's use the same dataset to demonstrate different plots:

```
# Comprehensive example with same dataset
np.random.seed(42)
# Create sample dataset: City population vs. average income
population = np.random.normal(500000, 200000, 50)  # City population
income = 30000 + 0.01*population + np.random.normal(0, 5000, 50)  # Avg income
income_error = np.random.uniform(1000, 3000, 50)  # Income measurement errors

fig = plt.figure(figsize=(16, 12))

# 1. Line Plot (if we consider population sorted)
plt.subplot(2, 3, 1)
sorted_idx = np.argsort(population)
plt.plot(range(len(population)), population[sorted_idx], 'o-', color='blue')
plt.title('Line Plot: Sorted Population Data')
plt.xlabel('City Index (sorted)')
plt.ylabel('Population')
plt.grid(True, alpha=0.3)

# 2. Scatter Plot
plt.subplot(2, 3, 2)
plt.scatter(population, income, alpha=0.7, c='red', edgecolors='black')
plt.title('Scatter: Population vs Income')
plt.xlabel('Population')
plt.ylabel('Average Income ($)')
plt.grid(True, alpha=0.3)

# 3. Error Visualization
plt.subplot(2, 3, 3)
plt.errorbar(population, income, yerr=income_error, fmt='o',
             alpha=0.7, ecolor='orange', capsize=3)
plt.title('Error Bars: Income Uncertainty')
plt.xlabel('Population')
plt.ylabel('Average Income ($)')
plt.grid(True, alpha=0.3)

# 4. Contour Plot
plt.subplot(2, 3, 4)
# Create density estimation
xy = np.vstack([population, income])
z = gaussian_kde(xy)(xy)
plt.tricontourf(population, income, z, levels=15, cmap='Blues')
plt.colorbar(label='Density')
plt.title('Contour: Population-Income Density')
plt.xlabel('Population')
plt.ylabel('Average Income ($)')

# 5. Histogram - Population distribution
plt.subplot(2, 3, 5)
plt.hist(population, bins=15, alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram: Population Distribution')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)

# 6. Histogram - Income distribution
plt.subplot(2, 3, 6)
plt.hist(income, bins=15, alpha=0.7, color='purple', edgecolor='black')
plt.title('Histogram: Income Distribution')
plt.xlabel('Average Income ($)')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
